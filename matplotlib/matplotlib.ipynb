```python
%matplotlib inline
from matplotlib import pyplot
pyplot.rcParams['font.family'] = 'IPAPGothic'  # 日本語フォントの設定
```

```python
import matplotlib

import numpy
```

```python
# x軸の点を-2π <= x <= 2πの範囲とする
x = numpy.linspace(-numpy.pi * 2, numpy.pi * 2, 1024, endpoint=True)
# sin(x)の値
y_sin = numpy.sin(x)
```

```python
# グラフの描画
# 2πごとに同じ値をとる sinの場合は x = 3.14, 6.28....の時1になる 
pyplot.plot(x, y_sin)
pyplot.show()
```

```python
# 周期関数（sin, cos）の描画
pyplot.plot(x, y_sin)
y_cos = numpy.cos(x)
pyplot.plot(x, y_cos)
pyplot.show()
```

```python
# グラフサイズの変更, 単位はインチ
pyplot.figure(figsize=(11, 7))

pyplot.plot(x, y_sin)
pyplot.plot(x, y_cos)

pyplot.show()
```

```python
# 線の色と線幅の変更
pyplot.figure(figsize=(11, 7))

pyplot.plot(x, y_sin, color='blue', linewidth=2.0)
pyplot.plot(x, y_cos, color='red', linewidth=2.0)

pyplot.show()
```

```python
# x軸、y軸の変更
pyplot.figure(figsize=(11, 7))

pyplot.plot(x, y_sin, color='blue', linewidth=2.0)
pyplot.plot(x, y_cos, color='red', linewidth=2.0)

# x軸の最小値最大値を、xの最小値最大値の1.2倍にする
pyplot.xlim(x.min() * 1.2, x.max() * 1.2)
# y軸の最小値最大値を、yの最小値最大値の1.2倍にする
pyplot.ylim(y_sin.min() * 1.2, y_sin.max() * 1.2)

pyplot.show()
```

```python
# 軸の目盛の設定
pyplot.figure(figsize=(11, 7))

pyplot.plot(x, y_sin, color='blue', linewidth=2.0)
pyplot.plot(x, y_cos, color='red', linewidth=2.0)

pyplot.xlim(x.min() * 1.2, x.max() * 1.2)
pyplot.ylim(y_sin.min() * 1.2, y_sin.max() * 1.2)

# x軸の目盛設定、リストを与える
pyplot.xticks([-numpy.pi * 2, -numpy.pi, 0, numpy.pi, numpy.pi * 2])
# y軸の目盛設定、リストを与える
pyplot.yticks([-1, -0.5, 0, 0.5, 1])

pyplot.show()
```

```python
# 目盛のラベルの設定
pyplot.figure(figsize=(11, 7))

pyplot.plot(x, y_sin, color='blue', linewidth=2.0)
pyplot.plot(x, y_cos, color='red', linewidth=2.0)

pyplot.xlim(x.min() * 1.2, x.max() * 1.2)
pyplot.ylim(y_sin.min() * 1.2, y_sin.max() * 1.2)

# x軸の目盛設定、リストを与える
pyplot.xticks([-numpy.pi * 2, -numpy.pi, 0, numpy.pi, numpy.pi * 2],
             [r'$-2\pi$', r'$-\pi$', r'$0$', r'$\pi$', r'$2\pi$']) # 別名
# y軸の目盛設定、リストを与える
pyplot.yticks([-1, -0.5, 0, 0.5, 1],
             [r'$-1.0$', r'$-0.5$', r'$0.0$', r'$0.5$', r'$1.0$']) # 別名

pyplot.show()
```

```python
# 凡例の追加
pyplot.figure(figsize=(11, 7))

# 凡例で使用するラベルを追加
pyplot.plot(x, y_sin, color='blue', linewidth=2.0, label=r'$\sin(x)$')
pyplot.plot(x, y_cos, color='red', linewidth=2.0, label=r'$\cos(x)$')

pyplot.xlim(x.min() * 1.2, x.max() * 1.2)
pyplot.ylim(y_sin.min() * 1.2, y_sin.max() * 1.2)

pyplot.xticks([-numpy.pi * 2, -numpy.pi, 0, numpy.pi, numpy.pi * 2],
              [r'$-2\pi$', r'$-\pi$', r'$0$', r'$\pi$', r'$2\pi$'])
pyplot.yticks([-1, -0.5, 0, 0.5, 1],
              [r'$-1.0$', r'$-0.5$', r'$0.0$', r'$0.5$', r'$1.0$'])

# 凡例表示
pyplot.legend()

pyplot.show()
```

```python
# タイトルの追加
pyplot.figure(figsize=(11, 7))

# 凡例で使用するラベルを追加
pyplot.plot(x, y_sin, color='blue', linewidth=2.0, label=r'$\sin(x)$')
pyplot.plot(x, y_cos, color='red', linewidth=2.0, label=r'$\cos(x)$')

pyplot.xlim(x.min() * 1.2, x.max() * 1.2)
pyplot.ylim(y_sin.min() * 1.2, y_sin.max() * 1.2)

pyplot.xticks([-numpy.pi * 2, -numpy.pi, 0, numpy.pi, numpy.pi * 2],
              [r'$-2\pi$', r'$-\pi$', r'$0$', r'$\pi$', r'$2\pi$'])
pyplot.yticks([-1, -0.5, 0, 0.5, 1],
              [r'$-1.0$', r'$-0.5$', r'$0.0$', r'$0.5$', r'$1.0$'])

# 凡例表示
pyplot.legend()

# タイトル追加
pyplot.title('Test Plot')

pyplot.show()
```
