# 機械学習
機械に人間のような学習能力を持たせる  
何十年、何百年と蓄積された気象データや株価のデータなど様々なデータを用い、パターンや構造を発見し有益な知識に変えることができる  

# 3つの手法
1. 教師あり学習
2. 教師なし学習
3. 強化学習

## 教師あり学習
データが正解ラベルを含むようなデータからパターンやルールを学習すること
正解ラベルが離散値か連続値かで2つのタスクに分けることができる
1. 分類
2. 回帰

### 分類
正解ラベルが離散値である教師あり学習
例：画像の分類（ある画像に含まれている物体が猫かどうか）
→ 猫の画像 と 猫でない画像 の2つに分けて、機械学習アルゴリズムで猫の特徴パターンを発見する
二値分類という（複数の場合は多クラス分類）
`境界線を見つける`

### 回帰
正解ラベルが連続値である教師あり学習
例：ソシャゲのイベントでユーザが何円課金するか
→ 年齢やログイン時間、直近の課金額（説明変数）などから次いくら課金するか（目的変数）を予測
`直線を見つける`


## 教師なし学習
正解ラベルを含まないデータから構造を発見し有益な知識を見つける
例：ソシャゲ内で似たようなユーザーの集まりを発見  
この場合は正解ラベルは与えられなく、事前にユーザーがどのようなグループに属するのかは不明
機械学習アルゴリズムを適用することで、あるグループのユーザーは常に課金してくれる、別のグループは全く課金してくれないと言った発見ができる
`非構造データからグループ（クラスタ）として構造を発見するタスク`をクラスタリングという


## 強化学習
エージェントがとる行動とその行動に対する報酬の組がデータとして与えられ、  
どのような行動を取るか決定するポリシー関数（政策関数）と呼ばれる関数を求める  
入力に対する明確な答えがない
例：将棋や囲碁をとくプログラム
→ 駒や碁石の次の動きを決定するためのポリシーを過去のデータから求めている




# 機械学習を利用するシステムの流れ
1. 前処理
2. 学習
3. 予測

## 前処理
学習データを整形する。
欠損値を補完したり、値のスケール（大きさ）を揃える
有益なデータを用意できなければ意味がない

## 学習（訓練）
機械学習アルゴリズムの選択をし、実際に予測モデルを作る
今のところあらゆるタスクに万能な機械学習アルゴリズムは存在しない

学習工程では、タスクを最も良く解くアルゴリズムを選定するために何度も予測モデルの評価を行う
どのような評価指標を用いるのかも重要

## 予測
良さそうな予測モデルが手に入ると、未知データに対してその予測モデルを用いて予測を行う。
未知データに学習時と同じ前処理を行うことを忘れないようにする
